### To get the SDK session object, you need to use the ophys_experiment_id. 
You can get an ophys_experiment_id in two ways. Each sessions has a unique behavior_session_id. Each OPHYS session has a unique ophys_session_id.  Each scientifica session has a single ophys_experiment_id. Mesoscope sessions will have multiple ophys_experiment_ids for each ophys_session_id. To conver between these, we wrote some useful functions:

```python
bsid = 855866701
# From behavior_session_id
ophys_experiment_id = sdk_utils.get_ophys_experiment_id_from_bsid(bsid,cache)
# From ophys_session_id
ophys_session_id = sdk_utils.get_ophys_session_id_from_bsid(bsid,cache)
ophys_experiment_id = sdk_utils.get_ophys_experiment_id_from_ophys_session_id(ophys_session_id,cache)
```

### Once you have the ophys_experiment_id, you can load the SDK session object by:
